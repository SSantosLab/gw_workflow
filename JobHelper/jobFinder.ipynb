## This file looks at the submitted DAGfiles, and compares them to the previously submitted jobsubs, and creates a list of the jobsubs left to be created

```python
def MakeJob(expNum):
    return "jobsub_submit_dag -G des --role=desgw --need-storage-modify /des/persistent/gw/exp --need-storage-modify /des/persistent/gw/forcephoto file://desgw_pipeline_{}.dag &> jobsub_{}.out &".format(expNum,expNum)
```

```python
dagFiles = np.loadtxt("dagFiles.txt",dtype=str) # This file is just a list of the dagmaker_1291889.out files for a given event
jobFiles = np.loadtxt("jobFiles.txt",dtype=str) # This file is just a list of the previously submitted jobsub_1291889.out files for a given event

showExposures = False # For debugging - use showExposures=True to view the exposure # string, which should be 7 chars
```

```python
for dagout in dagFiles:
    dag = str(dagout[9:-4]) # Only fetching the exposure number of the image
    noJob = True
    for file in jobFiles:
        if dag in file:
            # print("Found!")
            noJob=False
            break
    if noJob:
        if showExposures:
            print("Exposure #", dag)
        print(MakeJob(dag))
        print()
```
